```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
data=pd.read_csv("./cleaned_data.csv")
data
```

```python
Total_Transaction_Amount = data.groupby("CustomerId")["Amount"].sum().reset_index()
Total_Transaction_Amount.columns = ['CustomerId', 'Total_Transaction']


Average_Transaction_Amount = data.groupby("CustomerId")["Amount"].mean().reset_index()
Average_Transaction_Amount.columns = ['CustomerId', 'Average_Transaction']


Transaction_Count = data.groupby("CustomerId").size().reset_index(name='Transaction_Count')

# Calculate Standard Deviation of Transaction Amounts per customer
Standard_Deviation_Transaction_Amount = data.groupby("CustomerId")["Amount"].std().reset_index()
Standard_Deviation_Transaction_Amount.columns = ['CustomerId', 'Standard_Deviation']

# Merge all DataFrames on 'CustomerId'
Aggregate_Features = pd.merge(Total_Transaction_Amount, Average_Transaction_Amount, on='CustomerId')
Aggregate_Features = pd.merge(Aggregate_Features, Transaction_Count, on='CustomerId')
Aggregate_Features = pd.merge(Aggregate_Features, Standard_Deviation_Transaction_Amount, on='CustomerId')

# Display the aggregated features
Aggregate_Features.head()
```

```python
data["TransactionStartTime"]=pd.to_datetime(data["TransactionStartTime"])
```

```python
data["Hour"]=data["TransactionStartTime"].dt.hour
data["day"]=data["TransactionStartTime"].dt.day
data["week"]=data["TransactionStartTime"].dt.weekday
data["Month"]=data["TransactionStartTime"].dt.month
data["year"]=data["TransactionStartTime"].dt.year
data.head()
```

```python
data_1=data.copy()
data_1.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
# Identify categorical columns
categorical_columns = data_1.select_dtypes(include=['object']).columns.tolist()

# Create an instance of OneHotEncoder with sparse output
encoder = OneHotEncoder(sparse_output=True)  # Sparse matrix to save memory

# Fit and transform the data
one_hot_encoded = encoder.fit_transform(data_1[categorical_columns])

# Convert to DataFrame if necessary
one_hot_df = pd.DataFrame.sparse.from_spmatrix(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))

print(one_hot_df)
```

```python
num_cols = ["int64", "float64"]
numerical_cols = []
num_col=pd.DataFrame({})
for col in data.columns:
    if data[col].dtype.name in num_cols:  # Access the dtype of the column correctly
        numerical_cols.append(col)
        num_col[col]=data[col]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in data_1.columns:
    if col not in numerical_cols:
        data_1[col] = label_encoder.fit_transform(data_1[col])
data_1.head()
```
